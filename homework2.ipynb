```python
import heapq
```

```python
distance = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
```

```python
def heuristic(visited, n):

    # số thành phố chưa đi

    return (n - len(visited))
```

```python
def astar_tsp(distance):

    n = len(distance)

    start = 0

    open_list = []

    heapq.heappush(
        open_list,
        (0, start, [start])
    )

    best_cost = float('inf')

    best_path = None


    while open_list:

        cost, current, path = heapq.heappop(open_list)

        if len(path) == n:

            total_cost = cost + distance[current][start]

            if total_cost < best_cost:

                best_cost = total_cost

                best_path = path + [start]

            continue


        for next_city in range(n):

            if next_city not in path:

                new_cost = cost + \
                           distance[current][next_city]

                h = heuristic(path, n)

                f = new_cost + h

                heapq.heappush(
                    open_list,
                    (f,
                     next_city,
                     path + [next_city])
                )


    return best_path, best_cost
```

```python
path, cost = astar_tsp(distance)

print("Duong di:", path)

print("Chi phi:", cost)
```
