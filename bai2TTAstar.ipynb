```python
import heapq
```

```python


graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('A', 1), ('D', 2), ('G', 5)],
    'C': [('A', 4), ('D', 1)],
    'D': [('B', 2), ('C', 1), ('G', 3)],
    'G': []
}
```

```python


heuristic = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 1,
    'G': 0
}
```

```python
def astar(graph, start, goal):

   

    open_list = []

    heapq.heappush(open_list, (0, start))

  
    closed = set()

    
    parent = {}

    g = {}

    g[start] = 0


    while open_list:

      

        current = heapq.heappop(open_list)[1]


      

        if current == goal:

            path = []

            while current in parent:

                path.append(current)

                current = parent[current]

            path.append(start)

            path.reverse()

            return path


        closed.add(current)


       
        for neighbor, cost in graph[current]:

            if neighbor in closed:
                continue



            new_g = g[current] + cost


            if neighbor not in g or new_g < g[neighbor]:

                g[neighbor] = new_g

                f = new_g + heuristic[neighbor]

                heapq.heappush(
                    open_list,
                    (f, neighbor)
                )

                parent[neighbor] = current


    return None
```

```python
start = 'A'

goal = 'G'

path = astar(graph, start, goal)

print("Duong di:", path)
```
